### Testing ML with New Data

```python
new_data = pd.DataFrame([
    {"age": 67, "bmi": 28.4, "glucose": 145.0, "hypertension": 1, "heart_disease": 0},
    {"age": 45, "bmi": 23.7, "glucose": 95.0,  "hypertension": 0, "heart_disease": 0},
    {"age": 80, "bmi": 30.1, "glucose": 180.0, "hypertension": 1, "heart_disease": 1}
])

preds = pipe.predict(new_data)
probs = pipe.predict_proba(new_data)[:, 1]

results = new_data.copy()
results["predicted_stroke"] = preds
results["stroke_probability"] = probs

print(results)
```

### Model behavior and logic

The  model is doing what a good stroke risk model should:

- Assigning higher probabilities to older patients and those with hypertension, high glucose, or heart disease.

- Assigning lower probabilities to younger, healthier individuals.

This shows that the logistic regression coefficients likely have the expected signs:

- Positive weights for age, glucose, hypertension, and heart disease

- nNear-neutral or weak influence from BMI
