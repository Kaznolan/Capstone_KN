## Detect and handle outliers (IQR or z-score) - Life Expectancy Data

```python
# 
Q1 = le_merged['stroke'].quantile(0.25)
Q3 = le_merged['stroke'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"IQR = {IQR:.2f}, Lower bound = {lower_bound:.2f}, Upper bound = {upper_bound:.2f}")

# Identify outliers
outliers = le_merged[(le_merged['stroke'] < lower_bound) | (le_merged['stroke'] > upper_bound)]
print(f"Number of outliers: {len(outliers)}")

# Option 1: Remove outliers
df_no_outliers = le_merged[(le_merged['stroke'] >= lower_bound) & (le_merged['stroke'] <= upper_bound)]

# Option 2: Cap them instead of removing
df_capped = le_merged.copy()
df_capped['stroke'] = df_capped['stroke'].clip(lower=lower_bound, upper=upper_bound)
```

| Term                         | Meaning                                                                            | Interpretation                                                                                                        |
| ---------------------------- | ---------------------------------------------------------------------------------- | --------------------------------------------------------------------------------------------------------------------- |
| **IQR (7.56)**               | The spread of the middle 50% of your data (between the 25th and 75th percentiles). | Most stroke values lie within a 7.56-unit range.                                                                      |
| **Lower bound (-10.58)**     | The cutoff below which any value is considered an outlier.                         | Since stroke rates can’t be negative, effectively **no values below 0** will be valid.                                |
| **Upper bound (19.66)**      | The cutoff above which a value is considered an outlier.                           | Any country-year with a stroke rate **> 19.66 per 100,000** is flagged as an outlier.                                 |
| **Number of outliers (182)** | Count of data points exceeding these thresholds.                                   | There are 182 rows with stroke rates above 19.66 — very likely the high-rate countries like China, India, and Russia. |


The IQR range (7.56) shows most data are tightly clustered below 20.

But a few extreme values (e.g., China with 900+) are very far above normal.

The formula marks those 182 rows as outliers because they lie outside the normal range (upper bound = 19.66).

### Decisions for managing outliers in life Expectancy data

Those high outliers are not necessarily errors — they represent real health disparities between countries (e.g., China vs Oceania).


It might be better to detect and handle outliers within each continent separately, so the high rates founf in Asia and Russia the rest of the world data.


## Detect & Handle Outliers per Continent (IQR Method)

```python
# Create an empty list to store cleaned dataframes
cleaned_le = []

# Loop through each continent separately
for cont, group in le_merged.groupby('continent'):
    Q1 = group['stroke'].quantile(0.25)
    Q3 = group['stroke'].quantile(0.75)
    IQR = Q3 - Q1

    # Compute continent-specific bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Detect outliers
    outliers = group[(group['stroke'] < lower_bound) | (group['stroke'] > upper_bound)]
    print(f"🌍 {cont}: IQR={IQR:.2f}, lower={lower_bound:.2f}, upper={upper_bound:.2f}, outliers={len(outliers)}")

    # Option 1️⃣: Remove outliers for this continent
    group_no_outliers = group[(group['stroke'] >= lower_bound) & (group['stroke'] <= upper_bound)]

    # Option 2️⃣: Cap extreme values instead of removing
    # group['stroke'] = group['stroke'].clip(lower=lower_bound, upper=upper_bound)

    cleaned_le.append(group_no_outliers)

# Combine all cleaned groups back into one dataset
le_merged_clean = pd.concat(cleaned_le, ignore_index=True)

print(f"\n✅ Original size: {len(le_merged)}, Cleaned size: {len(le_merged_clean)}")
```
