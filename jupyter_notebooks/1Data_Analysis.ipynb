### Bar chart showing missing Values

```python
# ---  Calculate missing and non-missing counts per column ---
missing_counts = df_stroke.isna().sum()
non_missing_counts = df_stroke.notna().sum()

# ---  Combine into one DataFrame ---
missing_summary = pd.DataFrame({
    'Present': non_missing_counts,
    'Missing': missing_counts
})

# ---  Create stacked bar chart ---
ax = missing_summary.plot(
    kind='bar',
    stacked=True,
    figsize=(10, 6),
    color=['skyblue', 'salmon'],
    edgecolor='black'
)

# ---  Add count labels on each bar segment ---
for i, col in enumerate(missing_summary.index):
    present = missing_summary.loc[col, 'Present']
    missing = missing_summary.loc[col, 'Missing']
    
    # Label for 'Present' part
    ax.text(i, present / 2, str(int(present)),
            ha='center', va='center', fontsize=9, color='black', fontweight='bold')
    
    # Label for 'Missing' part (only if > 0)
    if missing > 0:
        ax.text(i, present + (missing / 2), str(int(missing)),
                ha='center', va='center', fontsize=9, color='black', fontweight='bold')

# ---  Formatting ---
plt.title('Count of Present vs Missing Values per Column', fontsize=14)
plt.xlabel('Columns', fontsize=12)
plt.ylabel('Number of Records', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Data Status', loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## Pie Chart of Gender distribution in the data 

```python
# Count gender values
gender_counts = df_stroke['gender'].value_counts()

# Create pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct='%1.1f%%',     # Show percentage
    startangle=90,         # Start pie at 90 degrees
    colors=['lightblue', 'lightcoral', 'lightgreen'],
    shadow=True
)

plt.title('Gender Distribution in Stroke Dataset', fontsize=14, fontweight='bold')
plt.show()
```

The pie chart shows that there is slightly more data for women in this data set than there is for men. Any gender-related analysis (like stroke rates by gender) should consider this imbalance.

This will need to be assessed after cleaning to see if anly steps are needed to balance the datset to prevent misleading insights based on gender.

## Detailed Analysis of missing data

```python
# Count missing values
missing_data = df_life_expectancy.isnull().sum()

# Calculate % missing
missing_percent = (missing_data / len(df_life_expectancy)) * 100

# Combine into a DataFrame
missing_summary = pd.DataFrame({
    'Missing Values': missing_data,
    'Missing (%)': missing_percent.round(2)
})

# Show only columns with missing data, sorted by % descending
missing_summary = missing_summary[missing_summary['Missing Values'] > 0].sort_values(by='Missing (%)', ascending=False)

# Display
print("ðŸ” Missing Data Summary:")
display(missing_summary)
```

## Number of strokes by Country

```python

# Standardize column names
df_life_expectancy.columns = [col.strip().title() for col in df_life_expectancy.columns]

# Group and sum strokes by country
strokes_by_country = (
    df_life_expectancy.groupby('Country')['Stroke']
    .sum()
    .sort_values(ascending=False)
    .head(20)  # Top 20 countries
)

# Generate unique colors for each bar
colors = plt.cm.tab20(np.linspace(0, 1, len(strokes_by_country)))

# Create the bar chart
plt.figure(figsize=(12, 6))
plt.bar(strokes_by_country.index, strokes_by_country.values, color=colors, edgecolor='black')
plt.xticks(rotation=75, ha='right')
plt.title('Top 20 Countries by Number of Strokes', fontsize=14, fontweight='bold')
plt.xlabel('Country', fontsize=12)
plt.ylabel('Number of Strokes', fontsize=12)
plt.tight_layout()
plt.show()
```
